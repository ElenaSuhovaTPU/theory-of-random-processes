2) Построить график винеровского процесса для моментов t из интервала 0 ≤ t ≤ 4 года с шагом h = 4x10^(-3)


```python
a,b = 0,4
t = []
n = 1000
h = 4/1000
for i in range(n):
  t+=[(h*i)]
t+=[4]
```
